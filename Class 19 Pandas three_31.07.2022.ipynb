```python
df_sales = pd.read_csv('sales_data_final.csv')
```

```python
df_sales.head()
```

```python
def profit_flag(a) :
    if a >= 0:
        return 'positive'
    else:
        return 'negative'
```

```python
profit_flag(55)
```

```python
profit_flag(-35)
```

```python
df_sales['flag_profit'] = df_sales['profit'].apply(profit_flag)
```

```python
df_sales.head()
```

```python
df_sales[df_sales['flag_profit'] == 'positive']
```

```python
df_sales['len_cust_name'] = df_sales['customer_name'].apply(len)
```

```python
df_sales.head()
```

```python
def quantity_flag(a) :
    if a < 5:
        return 'low'
    elif a> 5 and a <10:
        return 'medium'
    else:
        return 'high'
```

```python
df_sales['seg_quantity'] = df_sales['quantity'].apply(quantity_flag)
```

```python
df_sales.head()
```

```python
df_sales['square_qty'] = df_sales['quantity'].apply(lambda a : a**2) 
```
